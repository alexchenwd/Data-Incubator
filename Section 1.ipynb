```python
frame.modified_year = frame.modified_year.astype(int)
frame['OFFICE'] = frame['OFFICE'].apply(lambda x: x.replace('--', '') if '--' in x else x)

#only target for representatives
data5 = frame[frame.OFFICE.str.startswith('HON.')]

#find office with at least 5 members every year
pickoffice1 = frame.loc[:, ['OFFICE','PAYEE','CATEGORY','modified_year']]
pickoffice1 = pickoffice1[pickoffice1.CATEGORY == 'PERSONNEL COMPENSATION']
pickoffice1 = pickoffice1.groupby(['OFFICE','modified_year'])['PAYEE'].nunique()


member5office = []
for i in pickoffice1.index.get_level_values(0).unique():
    if (pickoffice1[i]>=5).all():
        member5office.append(i)

#find office at least serve 4+ years
d = []
for year in range(2011, 2017):
    serve4years = frame[frame.modified_year == year-4].OFFICE.unique()
    d.append(data5[(data5.modified_year == year) & (data5.OFFICE.isin(serve4years))])

data5 = pd.concat(d)
data5 = data5[data5.OFFICE.isin(member5office)]
data5
```

```python
q6 = data5[data5.CATEGORY == 'PERSONNEL COMPENSATION'].loc[:,['OFFICE','PAYEE','modified_year']]
q6.head(10)
        
```

```python
staff_member = q6.groupby(['OFFICE','modified_year'])['PAYEE'].unique()
```

```python
turnover = []
for member in staff_member.index.get_level_values(0).unique():
    for year in staff_member[member].index[0:-1]:
        if year+1 in staff_member[member].index:
            orignum = len(staff_member[member, year])
            leavenum = len(np.setdiff1d(staff_member[member, year].astype(str), staff_member[member, year+1].astype(str)))
            turnover.append(leavenum/orignum)
    
```

```python
np.median(turnover)
```
